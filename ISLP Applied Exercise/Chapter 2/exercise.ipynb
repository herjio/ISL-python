```python
import pandas as pd
import matplotlib.pyplot as plt

college = pd.read_csv('College.csv')
college.head()
```

```python
college2 = pd.read_csv('College.csv', index_col=0)
college3 = college.rename({'Unnamed: 0': 'College'},
                          axis=1)
college3 = college3.set_index('College')
college = college3
college.head()
```

```python
college.describe()
```

```python
pd.plotting.scatter_matrix(college[['Top10perc', 'Apps', 'Enroll']]);
```

```python
fig, ax = plt.subplots(figsize=(10, 8))

college.boxplot(column='Outstate', by='Private',ax=ax);
```

```python
college['Elite'] = pd.cut(college['Top10perc'],
                          [0, 50, 100],
                          labels=['No', 'Yes'])

college['Elite'].value_counts()
```

Binning universities into 2 groups based on the `Top10perc`. Whether their class exceeds 50% coming from top 10% of high schools.

```python
college.boxplot(column='Outstate', by='Elite');
```
